It contains very detailed weather data about the city of `Seoul`. Feel free to try other cities as well. We need to convert the named list to a data frame so that we can use data frame operations to process the data. Below is a simple example, which you may implement your own way to convert it to a data frame.


````python
df_current_weather = pd.DataFrame()
````

````python
# Create empty lists to store the JSON Data
weather = []
visibility = []
temp = []
temp_min = []
temp_max = []
pressure = []
humidity = []
wind_speed = []
wind_deg = []
````

Now assign the values in the `json_result` list into different vectors


````python
# # $weather is also a list with one element, its $main element indicates the weather status such as clear or rain
weather.append(response_json["weather"][0]["main"])
# # Get Visibility
visibility.append(response_json['visibility'])
# # Get current temperature 
temp.append(response_json['main']['temp'])
# # Get min temperature 
temp_min.append(response_json['main']['temp_min'])
# # Get max temperature 
temp_max.append(response_json['main']['temp_max'])
# # Get pressure
pressure.append(response_json['main']['pressure'])
# # Get humidity
humidity.append(response_json['main']['humidity'])
# # Get wind speed
wind_speed.append(response_json['wind']['speed'])
# # Get wind direction
wind_deg.append(response_json['wind']['deg'])
````

````python
wind_deg
````

Combine all vectors as columns of a data frame


````python
# Write Lists to DataFrame
df_current_weather['weather'] = weather
df_current_weather['visibility'] = visibility
df_current_weather['temperature'] = temp
df_current_weather['temperature_min'] = temp_min
df_current_weather['temperature_max'] = temp_max
df_current_weather['pressure'] = pressure
df_current_weather['humidity'] = humidity
df_current_weather['wind_speed'] = wind_speed
df_current_weather['wind_degree'] = wind_deg
````

````python
# Check the generated data frame
print(df_current_weather)
````

### Seoul

````python
# need to be replaced by your real API key
your_api_key = "277a72f5a19ff756ca573d7eaf5e56d9"
# Input `q` is the city name
city = "Seoul"
# Input `appid` is your API KEY, 
appid = your_api_key
# Input `units` are preferred units such as Metric or Imperial
units = "&units=metric"
#current_query <- list(q = "Seoul", appid = your_api_key, units="metric")
````

````python
forecast_response_json = requests.get(forecast_weather_url+"q="+city+"&appid="+appid).json()
````

````python
forecast_response_json
````

````python
df_predictions_seoul = pd.DataFrame()
````

````python
# Create empty lists to store the JSON Data
prediction_num = 0
list_prediction_num = []
city = []
weather = []
visibility = []
temp = []
temp_min = []
temp_max = []
pressure = []
humidity = []
wind_speed = []
wind_deg = []
````

````python
for num_forecasts in forecast_response_json['list']:
    df_predictions_seoul['prediction_num'] = prediction_num
    list_prediction_num.append(prediction_num)
    city.append(forecast_response_json['city']['name'])
    weather.append(forecast_response_json['list'][prediction_num]['weather'][0]['main'])
    visibility.append(forecast_response_json['list'][prediction_num]['visibility'])
    temp.append(forecast_response_json['list'][prediction_num]['main']['temp'])
    temp_min.append(forecast_response_json['list'][prediction_num]['main']['temp_min'])
    temp_max.append(forecast_response_json['list'][prediction_num]['main']['temp_max'])
    pressure.append(forecast_response_json['list'][prediction_num]['main']['pressure'])
    humidity.append(forecast_response_json['list'][prediction_num]['main']['humidity'])
    wind_speed.append(forecast_response_json['list'][prediction_num]['wind']['speed'])
    wind_deg.append(forecast_response_json['list'][prediction_num]['wind']['deg'])
    prediction_num += 1
````
